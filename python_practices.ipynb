```python
import math
import cmath # Import cmath for complex number operations

print("--- Quadratic Equation Solver (ax^2 + bx + c = 0) ---")

# Input coefficients
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculate the discriminant
d = (b**2) - (4*a*c)

print(f"\nThe discriminant is: {d}")

# Check if roots are real or complex for cleaner output
if d >= 0:
    # Calculate real roots
    sol1 = (-b - math.sqrt(d)) / (2*a)
    sol2 = (-b + math.sqrt(d)) / (2*a)
    print(f"The real roots are: {sol1:.2f} and {sol2:.2f}")
else:
    # Calculate complex roots using cmath
    sol1 = (-b - cmath.sqrt(d)) / (2*a)
    sol2 = (-b + cmath.sqrt(d)) / (2*a)
    print(f"The complex roots are: {sol1} and {sol2}")
```

```python
import math

num = int(input("Enter a number to check: "))

if num > 1:
    is_prime = True
    # We only need to check up to the square root of num
    for i in range(2, int(math.sqrt(num)) + 1):
        if (num % i) == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{num} is a prime number!")
    else:
        print(f"{num} is not a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# rite a Python Program to Print the Fibonacci sequence.
nterms = int(input("How many terms? "))
n1, n2 = 0,1
count = 0

if nterms <= 0:
  print("Please enter a positive integer")
elif nterms == 1:
    print("Fibonacci sequence upto", nterms, ":")
    print(n1)
else:
    print("Fibonacci sequence:")
    while count < nterms:
        print(n1)
        nth = n1 + n2
        n1 = n2
        n2 = nth
        count += 1
```

```python
# Write a Python Program to Check Armstrong Number?
num = int(input("enter a number: "))

# Convert to string to easily count digits and iterate
str_num = str(num)
num_digits = len(str_num)

sum_of_powers = 0

# Loop through each digit
for digit in str_num:
    sum_of_powers += int(digit) ** num_digits

# Check the result
if sum_of_powers == num:
    print(f"{num} is an Armstrong number!")
else:
    print(f"{num} is not an Armstrong number. (Sum is {sum_of_powers})")
```

```python
# Write a Python Program to Find Armstrong Number in an Interval.
lower = int(input("Enter the lower limit of the interval: "))
upper = int(input("Enter the upper limit of the interval: "))

for num in range(lower, upper + 1): # Iterate through the numbers
    order = len(str(num)) # Find the number of digits in 'num'
    temp_num = num
    sum = 0

    while temp_num > 0:
      digit = temp_num % 10
      sum += digit ** order
      temp_num //= 10

    if num == sum:
      print(num)
```
